```python
import pandas as pd
```

```python
# create column headers list
headers = ['code', 'Jan', 'Feb', 'March', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
```

```python
temp_avg2.head()
```

```python
# split code column in state and year
temp_avg2['State'] = temp_avg2['code'].str[0:3]
temp_avg2['Year'] = temp_avg2['code'].str[6:10]
```

```python
temp_avg2.head()
```

```python
# drop code column
temp_avg2.drop(['code'], axis=1, inplace=True)
temp_avg2.head()
```

```python
# convert State and year columns to int
temp_avg2 = temp_avg2.apply(pd.to_numeric)
```

```python
# move state and year columns to front
cols = temp_avg2.columns.tolist()
cols = cols[-2:] + cols[:-2]
temp_avg2 = temp_avg2[cols]
temp_avg2.head()
```
