```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug = "data/mouse_drug_data.csv"
clinical_trial = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse = pd.read_csv(mouse_drug)
trial = pd.read_csv(clinical_trial)

# Combine the data into a single dataset
comb_data = trial.merge(mouse, on = ["Mouse ID"])

# Display the data table for preview
comb_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_time = comb_data.groupby(["Drug","Timepoint"]).mean()["Tumor Volume (mm3)"]
# Convert to DataFrame
tumor_df = pd.DataFrame(tumor_time)
tumor_df2 = tumor_df.reset_index()
# Preview DataFrame
tumor_df2.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

sem_data = comb_data.groupby(["Drug","Timepoint"]).sem()["Tumor Volume (mm3)"]


# Convert to DataFrame
sem_df = pd.DataFrame(sem_data)
sem_df2 = sem_df.reset_index()

# Preview DataFrame

sem_df2.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

tumor_pretty = tumor_df2.pivot(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")
sem_pretty = sem_df2.pivot(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")
# Preview that Reformatting worked
tumor_pretty
#sem_pretty.head()
```

```python
# Generate the Plot (with Error Bars)

plt.figure(figsize=(20,10))

plt.errorbar(tumor_pretty.index, tumor_pretty["Capomulin"], yerr = sem_pretty["Capomulin"], fmt="*", ms =20, linestyle = "solid", capsize = 5)
plt.errorbar(tumor_pretty.index, tumor_pretty["Infubinol"], yerr = sem_pretty["Infubinol"], fmt="D", ms=10, linestyle = "solid", capsize = 5)
plt.errorbar(tumor_pretty.index, tumor_pretty["Ketapril"], yerr = sem_pretty["Ketapril"], fmt="o", ms=10, linestyle = "solid", capsize =5)
plt.errorbar(tumor_pretty.index, tumor_pretty["Placebo"], yerr = sem_pretty["Placebo"], fmt="s", ms=10,  linestyle ="solid", capsize =5)

plt.title("Tumor Volume Change in Response to Treatment")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time in Days")
plt.xticks(tumor_pretty.index)
plt.grid(color ="black", alpha = .5)
plt.legend(loc = "best", fontsize = "x-large", title = "Treatment Used")

plt.savefig("PymaceuticalsFig1.png")
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 

meta = comb_data.groupby(["Drug", "Timepoint"]).mean()["Metastatic Sites"]

# Convert to DataFrame

meta_df = pd.DataFrame(meta)
meta_df2 = meta_df.reset_index()

# Preview DataFrame
meta_df2.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 

meta_sem = comb_data.groupby(["Drug","Timepoint"]).sem()["Metastatic Sites"]

# Convert to DataFrame

meta_sem_df = pd.DataFrame(meta_sem)
meta_sem_df2 = meta_sem_df.reset_index()

# Preview DataFrame

meta_sem_df2.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

meta_RF = meta_df2.pivot(index = "Timepoint", columns = "Drug", values = "Metastatic Sites")
meta_sem_RF = meta_sem_df2.pivot(index = "Timepoint", columns = "Drug", values = "Metastatic Sites")

# Preview that Reformatting worked
meta_sem_RF.head()
```

```python
# Generate the Plot (with Error Bars)

plt.figure(figsize=(20,10))

plt.errorbar(meta_RF.index, meta_RF["Capomulin"], yerr = meta_sem_RF["Capomulin"], fmt="*", ms =20, linestyle = "solid", capsize =5)
plt.errorbar(meta_RF.index, meta_RF["Infubinol"], yerr = meta_sem_RF["Infubinol"], fmt="D", ms=10, linestyle = "solid", capsize = 5)
plt.errorbar(meta_RF.index, meta_RF["Ketapril"], yerr = meta_sem_RF["Ketapril"], fmt="o", ms=10, linestyle = "solid", capsize = 5)
plt.errorbar(meta_RF.index, meta_RF["Placebo"], yerr = meta_sem_RF["Placebo"], fmt="s", ms=10,  linestyle ="solid", capsize =5)

plt.title("Increase in Metastatic Tumors Over Course of Treatment")
plt.ylabel("Metastatic Sites")
plt.xlabel("Days of Treatment")
plt.xticks(tumor_pretty.index)
plt.grid(color ="black", alpha = .5)
plt.legend(loc = "best", fontsize = "x-large", title = "Treatment Used")

plt.savefig("PymaceuticalsFig2.png")
plt.show()


# Save the Figure

# Show the Figure
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

alive= comb_data.groupby(["Drug", "Timepoint"]).count()["Mouse ID"]

# Convert to DataFrame

alive_df = pd.DataFrame(alive)
alive_df2 = alive_df.reset_index()
alive_df2 = alive_df2.rename(columns = {"Mouse ID": "Mice Left Alive"})

# Preview DataFrame
alive_df2.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

alive_RF = alive_df2.pivot(index = "Timepoint", columns = "Drug", values = "Mice Left Alive")

# Preview the Data Frame
alive_RF.head()
```

```python
# Generate the Plot (Accounting for percentages)

plt.figure(figsize=(20,10))

plt.plot(alive_RF.index, alive_RF["Capomulin"]*4, marker = "*", ms =20)
plt.plot(alive_RF.index, alive_RF["Infubinol"]*4, marker="D", ms=10)
plt.plot(alive_RF.index, alive_RF["Ketapril"]*4, marker="o", ms=10)
plt.plot(alive_RF.index, alive_RF["Placebo"]*4, marker="s", ms=10)

plt.title("Survival Rate of Mice Over Course of Treatment")
plt.ylabel("Percent of Mice Still Alove")
plt.xlabel("Days of Treatment")
plt.xticks(tumor_pretty.index)
plt.grid(color ="black", alpha = .5)
plt.legend(loc = "best", fontsize = "x-large", title = "Treatment Used")

plt.savefig("PymaceuticalsFig3.png")
plt.show()

# Save the Figure

# Show the Figure
```

Instead, I should have found the average tumor volume on day 45 for all surviving mice.  I did the correct calculation below, but ran out of time to do the graph. 

```python
# Calculate the percent changes for each drug
summ = comb_data.groupby(["Drug", "Mouse ID", "Timepoint"]).mean()
sum_df = pd.DataFrame(summ)
sum_df2 = sum_df.reset_index()
sumC = sum_df2[sum_df2["Drug"]=="Capomulin"]
sumCRF = sumC.pivot(index = "Timepoint", columns = "Mouse ID", values = "Tumor Volume (mm3)")

C_end =[]
time = [45,40,35,30,25,20,15,10,5,0]
for column in sumCRF:
    for row in time:
        if sumCRF.loc[row, column] >= 0:
            end_CV = sumCRF.loc[row, column]
            C_end.append(end_CV)
            break
            
C_avg = sum(C_end)/len(C_end)

sumI = sum_df2[sum_df2["Drug"]=="Infubinol"]
sumIRF = sumI.pivot(index = "Timepoint", columns = "Mouse ID", values = "Tumor Volume (mm3)")

I_end =[]
time = [45,40,35,30,25,20,15,10,5,0]
for column in sumIRF:
    for row in time:
        if sumIRF.loc[row, column] >= 0:
            end_IV = sumIRF.loc[row, column]
            I_end.append(end_IV)
            break
            
I_avg = sum(I_end)/len(I_end)


sumK = sum_df2[sum_df2["Drug"]=="Ketapril"]
sumKRF = sumK.pivot(index = "Timepoint", columns = "Mouse ID", values = "Tumor Volume (mm3)")

K_end =[]
time = [45,40,35,30,25,20,15,10,5,0]
for column in sumKRF:
    for row in time:
        if sumKRF.loc[row, column] >= 0:
            end_KV = sumKRF.loc[row, column]
            K_end.append(end_KV)
            break
            
K_avg = sum(K_end)/len(K_end)


sumP = sum_df2[sum_df2["Drug"]=="Placebo"]
sumPRF = sumP.pivot(index = "Timepoint", columns = "Mouse ID", values = "Tumor Volume (mm3)")

P_end =[]
time = [45,40,35,30,25,20,15,10,5,0]
for column in sumPRF:
    for row in time:
        if sumPRF.loc[row, column] >= 0:
            end_PV = sumPRF.loc[row, column]
            P_end.append(end_PV)
            break
            
P_avg = sum(P_end)/len(P_end)

the_end = (C_avg, I_avg, K_avg, P_avg)
the_end
```
