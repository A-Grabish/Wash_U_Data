### Analysis
1.   Urban cities accounted for over 60% of total rides taken and total revenue, however urban cities had the lowest average fare per ride with rural rides having the highest average fare per ride.

2.   Urban cities had over 80% of all total drivers with a total of 2405 drivers.  However, these urban drivers only accepted 1625 rides, meaning there were 780 more urban drivers than urban rides taken.  Stated another way, there were 1.48 urban drivers per urban ride taken

3.   Due to these facts, urban drivers had the lowest revenue per driver.  The average suburban driver received more than twice the revenue as the average urban driver, while the average rural driver received more than three times the revenue of the average urban driver.  This suggests the urban markets in general are over saturated with drivers, and rural markets have the potential to grow there respective driver pools.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pathC ="data/city_data.csv"
pathR = "data/ride_data.csv"
city_data = pd.read_csv(pathC)
ride_data = pd.read_csv(pathR)
```

```python
total_data = ride_data.merge(city_data)
total_data = total_data.set_index(['city'])
total_data.head()
```

```python
urban = total_data[total_data["type"] == "Urban"]
suburban = total_data[total_data["type"] == "Suburban"]
rural = total_data[total_data["type"] == "Rural"]
```

```python
#avg fare
urban_avg = urban.groupby(["city"]).mean()["fare"]
suburban_avg = suburban.groupby(["city"]).mean()["fare"]
rural_avg = rural.groupby(["city"]).mean()["fare"]

#rides per city
urban_rides = urban.groupby(["city"]).count()["date"]
suburban_rides = suburban.groupby(["city"]).count()["date"]
rural_rides = rural.groupby(["city"]).count()["date"]

#drivers per city
urban_drive = urban.groupby(["city"]).mean()["driver_count"]
suburban_drive = suburban.groupby(["city"]).mean()["driver_count"]
rural_drive = rural.groupby(["city"]).mean()["driver_count"]
```

```python
plt.figure(figsize = (12,9))
plt.scatter(urban_rides, urban_avg, s = urban_drive*100, alpha = .5, edgecolor = "black", color = "coral", label = "Urban")
plt.scatter(suburban_rides, suburban_avg, s = suburban_drive*100, alpha = .5, edgecolor = "black", color = "lightskyblue", label = "Suburban")
plt.scatter(rural_rides, rural_avg, s = rural_drive*100, alpha = .5, edgecolor = "black", color = "gold", label = "Rural")


import matplotlib.patches as mpatches
plt.title("Pyber Ride Data")
plt.xlabel("Total Rides per City")
plt.ylabel("Average Fare per City")
plt.grid(True)

coral_patch = mpatches.Patch(color='coral', label='Urban', alpha = .5)
blue_patch = mpatches.Patch(color = "lightskyblue", label = "Suburban", alpha = .5)
gold_patch = mpatches.Patch(color = "gold", label = "Rural", alpha = .5)
plt.legend(loc = "best", handles=[coral_patch, blue_patch, gold_patch], title = "City Type")

plt.text(43,32, "Note: Circle Size is Relative to \nDriver Count for Individual Cities")
plt.savefig("PyberFig1.png")
```

```python

#ride%
total_ride = total_data.count()["ride_id"]
urban_total_ride = urban.count()["ride_id"]
suburban_total_ride = suburban.count()["ride_id"]
rural_total_ride = rural.count()["ride_id"]
ride_data = [urban_total_ride/total_ride*100, suburban_total_ride/total_ride*100, rural_total_ride/total_ride*100]
plt.pie(ride_data, labels = ["Urban", "Suburban", "Rural"], autopct = "%1.1f%%", colors = ["lightcoral", "lightskyblue", "gold"], explode = [.1,0,0], shadow = True )
plt.axis("equal")
plt.title("% of Total Rides Taken by City Type")
plt.savefig("PyberFig2.png")
```

```python
#%of fares
fare_data = total_data.groupby(["type"]).sum()["fare"]
fare_total = total_data.sum()["fare"]
plt.pie(100*fare_data/fare_total, labels = ["Rural", "Suburban", "Urban"], shadow = True,
        colors = ["gold", "lightskyblue", "lightcoral"], autopct = "%1.1f%%", explode = [0,0,.1], startangle = 270)
plt.axis("equal")
plt.title("% of Total Revenue by City Type")
plt.savefig("PyberFig3.png")
```

```python
#% of total drivers
total_drive = urban_drive.sum() + suburban_drive.sum() + rural_drive.sum()
plt.pie([urban_drive.sum()/total_drive*100, suburban_drive.sum()/total_drive*100, rural_drive.sum()/total_drive*100],
       autopct = "%1.1f%%", colors = ["lightcoral", "lightskyblue", "gold"], explode = [.1,0,0], shadow = True,
       labels = ["Urban", "Suburban", "Rural"])
plt.axis("equal")
plt.title("% of Drivers by City Type")
plt.savefig("PyberFig4.png")
```
